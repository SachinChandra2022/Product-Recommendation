```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500000, stop_words='english')
```

```python
cv.fit_transform(dataset_df['tags']).toarray()
```

```python
vectors = cv.fit_transform(dataset_df['tags']).toarray()
```

```python
len(cv.get_feature_names())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
dataset_df["tags"]=dataset_df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(Product_Category):
    product_index=dataset_df[dataset_df['Product_Category']==Product_Category].index[0]
    distances=similarity[product_index]
    products=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in products:
        print(dataset_df.iloc[i[0]].Product_Category)
```

```python
recommend('Home')
```
